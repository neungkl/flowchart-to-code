```python
import numpy as np
import pandas as pd
import os

import matplotlib
import matplotlib.pyplot as plt

from PIL import Image

%matplotlib inline
```

```python
os.getcwd()
```

```python
data_dir = os.path.join(os.getcwd(), "data")
```

```python
sample_id = "1-1"
picture = os.path.join(data_dir, "sample-" + sample_id + ".jpg")

image = Image.open(picture).convert('L')
image = image.convert()
image = np.asarray(image) / 255
```

```python
pos = pd.read_csv(os.path.join(data_dir, "sample-" + sample_id + "-block-pos.csv"), header=None)
pos = pos[[1,2]].values
pos
```

```python
plt.figure(figsize=(5,10))
plt.scatter(pos[:,0], pos[:,1], c='r')
plt.imshow(image, cmap="gray")
plt.show()
```
